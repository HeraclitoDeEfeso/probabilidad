Lo siguiente que haremos será crear una muestra de 50 números con base en una distribución binomial con parámetros correspondientes a 10 experimentos Bernoulli y una probabilidad de éxito de 0,3. Ahora simulamos una segunda muestra de 50 dotas que tendrán por base la distribución empírica que resulta de la muestra anterior. Calculamos también su media y varianza muestral además de graficar el histograma.

A continuación repetimos nuestra experimento dos veces más y presentamos todos los datos juntos.

```python
bino = binomial(50, 10, 0.3)
emp = empirica(50, bino)
print("Muestra binomial - ", end="")
estadist(bino)
histograma(bino, [1])
print("Muestra empírica - ", end="")
estadist(emp)
histograma(emp, [1])
```
